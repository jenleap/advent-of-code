How many total square feet of wrapping paper should the elves order?

```python
total_paper = 0

with open("input.txt", "r") as f:
    for present in f:
        l,w,h = map(int, present.strip().split('x'))

        sides = [l * w, w * h, l * h]
        total_paper += 2 * sum(sides) + min(sides)

print(total_paper)

```

How many total feet of ribbon should the elves order?

```python
total_ribbon = 0

with open("input.txt", "r") as f:
    for present in f:
        l,w,h = map(int, present.strip().split('x'))
        sides = [l,w,h]
        bow = l*w*h
        sides.remove(max(sides))
        perimeter = (2 * sides[0]) + (2 * sides[1])
        total_ribbon += bow + perimeter

print(total_ribbon)
```
