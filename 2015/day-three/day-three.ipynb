How many houses receive at least one present?

```python
houses = {}
position = [0,0]

with open("input.txt", "r") as f:
    directions = f.read()

for dir in directions:
    if (dir == "v"):
        position[1] = position[1] + 1
    elif (dir == "^"):
        position[1] = position[1] - 1
    elif (dir == ">"):
        position[0] = position[0] + 1
    elif (dir == "<"):
        position[0] = position[0] - 1
    else:
        print(f"{dir} is not supported")

    house = "_".join(str(p) for p in position)

    if house in houses:
        houses[house] += 1
    else:
        houses[house] = 1

print(len(houses))
```

With Santa and Robo-Santa alternating, how many houses receive at least one present?

```python
houses = {}
santa_pos = [0,0]
robo_pos = [0,0]
counter = 0

def move_santa(position, dir):
    if (dir == "v"):
        position[1] = position[1] + 1
    elif (dir == "^"):
        position[1] = position[1] - 1
    elif (dir == ">"):
        position[0] = position[0] + 1
    elif (dir == "<"):
        position[0] = position[0] - 1
    else:
        print(f"{dir} is not supported")
    
    house = "_".join(str(p) for p in position)

    if house in houses:
        houses[house] += 1
    else:
        houses[house] = 1
    

with open("input.txt", "r") as f:
    directions = f.read()

for dir in directions:
    if (counter%2 == 0):
        move_santa(santa_pos, dir)
    else:
        move_santa(robo_pos, dir)

    counter += 1

print(len(houses))
```

```python
# Alternate solution

moves = {
    '^': (0, 1),
    'v': (0, -1),
    '>': (1, 0),
    '<': (-1, 0),
}

santa_pos = [0,0]
robo_pos = [0,0]

houses = {(0,0)}

with open("input.txt", "r") as f:
    directions = f.read()

for index, dir in enumerate(directions):
    pos = santa_pos if index%2 == 0 else robo_pos
    move_x, move_y = moves[dir]
    pos[0] += move_x
    pos[1] += move_y

    houses.add((pos[0], pos[1]))

print(len(houses))
```
