To what floor do the instructions take Santa?

```python
current_floor = 0

with open("input.txt", "r") as f:
    directions = f.read()

for floor in directions:
    if (floor == "("):
        current_floor += 1
    else:
        current_floor -= 1

print(current_floor)
```

What is the position of the character that causes Santa to first enter the basement?

```python
current_floor = 0

with open("input.txt", "r") as f:
    directions = f.read()

position = 1
for floor in directions:
    if (floor == "("):
        current_floor += 1
    else:
        current_floor -= 1
    
    if (current_floor < 0):
        break
    else:
        position += 1

print(position)
```
